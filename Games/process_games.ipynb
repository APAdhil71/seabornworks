```python
import seaborn as sns 
import pandas as pd
import matplotlib.pyplot as plt                                                                                                                  
```

```python
df = pd.read_csv("C:\\Users\\adhil\\OneDrive\\Desktop\\seabornworks\\Games\\games.csv")
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
sns.histplot(df["white_rating"])
plt.title("White Rating Distribution")
plt.show()
```

```python
sns.histplot(df["black_rating"])
plt.title("Black Rating Distribution")
plt.show()
```

```python
# Compare Two Numeric Columns (Scatter Plot)
sns.scatterplot(data=df, x="white_rating", y="black_rating")
plt.title("White Rating vs Black Rating")
plt.show()
```

```python
# Compare Category vs Numeric (Boxplot)

sns.boxplot(data=df, x="winner", y="turns")
plt.title("Winner vs Number of Turns")
plt.show()
```

```python
#  Compare Category vs Numeric (Bar Plot  Mean Values)
#  average turns by victory_status

sns.barplot(data=df, x="victory_status", y="turns")
plt.title("Average Turns by Victory Status")
plt.show()
```

```python
# Compare Frequency of Categories (Countplot)

# Example: number of games per winner

sns.countplot(data=df, x="winner")
plt.title("Winner Count")
plt.show()
```
