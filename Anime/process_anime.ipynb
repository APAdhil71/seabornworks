```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:\\Users\\adhil\\OneDrive\\Desktop\\seabornworks\\Anime\\anime.csv")
```

```python
df.info()
df.isnull().sum()
```

```python
df["synopsis"] = df["synopsis"].fillna("Unknown")
```

```python
df["start_date"] = df["start_date"].fillna("Unknown")
```

```python
df["end_date"] = df["end_date"].fillna("Unknown")
```

```python
df["episodes"] = df["episodes"].fillna(df["episodes"].median())
```

```python
# Boxplot â€” Compare Score by Type
sns.boxplot(data=df, x="type", y="score")
plt.title("Anime Score by Type")
plt.show()
```

```python
# Boxplot â€” Compare Members by Type
sns.boxplot(data=df, x="type", y="members")
plt.title("Anime Popularity Members by Type")
plt.show()
```

```python
# Barplot â€” Average Score for Each Type
sns.barplot(data=df, x="type", y="score", estimator='mean')
plt.title("Average Score by Anime Type")
plt.show()
```

```python
# Scatter Comparison â€” Score vs Members
sns.scatterplot(data=df, x="members", y="score", hue="type")
plt.title("Score vs Members by Anime Type")
plt.show()
```

```python
# Violin Plot â€” Score Distribution by Type
sns.violinplot(data=df, x="type", y="score")
plt.title("Score Distribution by Anime Type")
plt.show()
```

```python
sns.lineplot(data=df, x="rank", y="score")
plt.title("Rank vs Score (Line Plot)")
plt.xlabel("Rank (Lower = Better)")
plt.ylabel("Score")
plt.show()
```

```python
sns.scatterplot(data=df, x="popularity", y="score")
plt.title("Popularity vs Score")
plt.xlabel("Popularity (Lower = More Popular)")
plt.ylabel("Score")
plt.show()
```
